```python
import pandas as pd

import pandas as pd

# Assuming the file is uploaded to Colab's file system
file_path =  '/content/agstar-livestock-ad-database-combined.xlsx'

# Load the CSV file using semicolon as the delimiter
data = pd.read_excel(file_path)
```

```python
# Display the first few rows to verify
data
```

```python
# Convert values to float in specific columns
data['Electricity Generated (kWh/yr)'] = pd.to_numeric(data['Electricity Generated (kWh/yr)'], errors='coerce')
data['Electricity Generated (kWh/yr)'] = data['Electricity Generated (kWh/yr)'].astype(float)
```

```python
data = data.dropna(subset=['Electricity Generated (kWh/yr)'])

# Print the DataFrame after dropping NaN values
data
```

```python
data_sorted = data.sort_values(by='Year Operational')

# Print the sorted DataFrame
data_sorted
```

```python
# Filter the DataFrame to show only rows where 'Year Operational' is greater than 2010
filtered_data = data_sorted[data_sorted['Year Operational'] > 2010]

# Print the filtered DataFrame
filtered_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error

# Assuming 'data' contains your DataFrame with the relevant columns including 'Electricity Generated (kWh/yr)'

# Split the data into features (X) and target variable (y)
X = filtered_data[['Year Operational']]  # Features
y = filtered_data['Electricity Generated (kWh/yr)']  # Target variable

# Split the data into training and testing sets (60% train, 40% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Initialize the Random Forest Regressor model
rf_model = RandomForestClassifier(n_estimators= 300, max_depth=7, random_state=42)

# Train the model on the training data
rf_model.fit(X_train, y_train)

train_score = rf_model.score(X_train, y_train)
test_score = rf_model.score(X_test, y_test)

print("Training score:", train_score)
print("Testing score:", test_score)
```
